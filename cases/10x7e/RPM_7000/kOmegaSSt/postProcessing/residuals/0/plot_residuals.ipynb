Plot Residuals

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker

# p target residuals: below 1e-4
# U target residuals: acceptable below 1e-4, good below 1e-5
# k target residuals: acceptable below 1e-4, good below 1e-5


# Read file
df = pd.read_csv("residuals.dat", sep="\t",skiprows=[0,2])

# Clean up column names
df.columns = df.columns.str.replace("#", "", regex=False).str.strip()

# Plot
plt.style.use("default")
fig, axs = plt.subplots(3, 1, figsize=(20, 14), sharex=True)

# --- pressure ---
axs[0].plot(df.iloc[:, 0], df["p"], label="p")
axs[0].set_yscale("log")
axs[0].set_ylim(1e-5, 1e-1)
axs[0].set_title("Pressure Residual", fontsize=18)
axs[0].set_ylabel("Residual", fontsize=16)
axs[0].legend(fontsize=14)
axs[0].grid(True, which="major")

# --- momentum ---
for c in ["Ux", "Uy", "Uz"]:
    axs[1].plot(df.iloc[:, 0], df[c], label=c)
axs[1].set_yscale("log")
axs[1].set_ylim(1e-5, 1e-1)
axs[1].set_title("Momentum Residuals", fontsize=18)
axs[1].set_ylabel("Residual", fontsize=16)
axs[1].legend(fontsize=14)
axs[1].grid(True, which="major")

# --- energy ---
axs[2].plot(df.iloc[:, 0], df["k"], label="k")
axs[2].set_yscale("log")
axs[2].set_ylim(1e-5, 1e-1)
axs[2].set_title("Energy Residual", fontsize=18)
axs[2].set_ylabel("Residual", fontsize=16)
axs[2].set_xlabel("Time", fontsize=16)
axs[2].legend(fontsize=14)
axs[2].grid(True, which="major")

for ax in axs:
    ax.tick_params(axis="both", labelsize=16)

plt.tight_layout()
plt.show()



```
